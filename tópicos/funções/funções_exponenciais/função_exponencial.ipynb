# Função exponencial

## Definição

Se $a \in \R^{*}_{+}$ e $a \neq 1$, diremos que $f$ é uma função exponencial se

$$f(x) = a^{x}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def exponential_f(function,
                  linspace=np.linspace(-10, 10, 1000),
                  xticks=range(-10,10+1, 2),
                  label=''):
    X = linspace
    Y = [function(x) for x in X]

    plt.axhline(0, color='black');
    plt.axvline(0, color='black');
    plt.plot(X, Y, label=label);
    plt.xticks(xticks);
    plt.xlabel('$x$');
    plt.ylabel('$y$');
    plt.legend();
```

## Função crescente ($a > 1$)

```python
exponential_f(lambda x: 2**x, label='$2^x$')
```

```python
print(2**(10))

print(2**(-10))
print(1/1024)
```

## Função decrescente ($0 < a < 1$)

```python
exponential_f(lambda x: 0.5**x, label='$0.5^x$')
```

## Visualizando ambas (crescente e decrescente)

```python
exponential_f(lambda x: 2**x, label='$2^x$')
exponential_f(lambda x: 0.5**x, label='$0.5^x$')
```

```python
print(2**10)
print(0.5**(10))
print(1/(2**(10)))
```

## O que ocorre quando $a \rightarrow 0$ ?

```python
exponential_f(lambda x: 0.1**x, label='$0.1^x$')
exponential_f(lambda x: 0.11**x, label='$0.11^x$')
exponential_f(lambda x: 0.12**x, label='$0.12^x$')
exponential_f(lambda x: 0.13**x, label='$0.13^x$')

plt.xlim(-10, -8);
```

```python
# quanto menor for o a, maior será o y para valores de x negativos
print(0.1**-10)
print(0.2**-10)
print(0.5**-10)
print(1**-10)
```

## O que ocorre com $a < 0$ ?
- Se $a < 0$, então para $a^x$ temos que $x$ não pode ser racional, pois então não temos solução em $\R$
- Também temos que $a$ elevado a expoente ímpar resulta em um valor ímpar: $(-2)^3 = -8$
- Já $a$ elevado a expoente par resulta em um valor par: $(-2)^2 = 4$

```python
a = -2
f = lambda x: (a)**x

X = range(-4,4+1,1)
Y = [f(x) for x in X]

plt.axhline(0, color='black', linewidth=1);
plt.axvline(0, color='black', linewidth=1);
plt.plot(X, Y, marker='o', label='$(-2)^x$');
plt.xticks(X);
plt.yticks(range(-8,16+1,3));
plt.xlabel('$x$');
plt.ylabel('$y$');
plt.legend();
```

## Por que $a \neq 1$?
1 elevado a qualquer valor resulta em 1

```python
exponential_f(lambda x: 1**x, label='$1^x$');
```

## Por que $a \neq 0$?
- 0 elevado a qualquer valor resulta em 0
- $0^0$ é uma indeterminação

```python
print(0.01**5)    # tende a 0^{+\infty} -> 0
print(0.01**-5)   # tende 0^{-\infty} -> \infty
```
