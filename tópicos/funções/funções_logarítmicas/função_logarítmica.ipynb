```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
b = 10
fx = lambda x: math.log(x, b)

x = np.linspace(0.001, 12, 100)
y = [fx(i) for i in x]

plt.plot(x, y, label='$y = \log_{' + str(b) + '}{x}$');
plt.scatter(1, 0, color='r', zorder=3);
plt.scatter(10, 1, color='r', zorder=3);

plt.axhline(0, color='black', linestyle='-', linewidth=1);
plt.axvline(0, color='black', linestyle='-', linewidth=1);
plt.vlines(x=1, ymin=-2, ymax=0, color='grey', linestyle='--');
plt.vlines(x=10, ymin=-2, ymax=1, color='grey', linestyle='--');
plt.hlines(y=1, xmin=-2, xmax=10, color='grey', linestyle='--');

plt.xlim(-2,12);
plt.ylim(-2,2);
plt.xticks(range(-2,13));

plt.xlabel('$x$');
plt.ylabel('$y$');
plt.legend(fontsize=12);
```

```python
def log_f(function, label,
          linspace=np.linspace(0.01, 10, 1000),
          xticks=range(0, 10+1)):
    f = function
    X = linspace
    Y = [f(x) for x in X]
    
    plt.plot(X, Y, label=label);
    plt.axhline(0, color='black');
    plt.axvline(0, color='black');
    plt.scatter(1, 0, color='r', zorder=2);
    plt.xlabel('$x$');
    plt.ylabel('$y$');
    plt.xticks(xticks);
    plt.legend();
```

```python
log_f(lambda x, b=10: math.log(x, b), label='$\log_{10}{x}$')
```

```python
log_f(lambda x, b=2: math.log(x, b), label='$\log_{2}{x}$')
log_f(lambda x, b=3: math.log(x, b), label='$\log_{3}{x}$')
log_f(lambda x, b=5: math.log(x, b), label='$\log_{5}{x}$')
log_f(lambda x, b=10: math.log(x, b), label='$\log_{10}{x}$')
```
