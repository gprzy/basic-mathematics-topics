```python
import numpy as np
import matplotlib.pyplot as plt

from sympy import symbols, solve

_x_, _y_ = symbols('x y')
```

$$
\begin{cases}
2x + 3y = 6 \\
4x + 2y = 8
\end{cases}
$$

Resolvendo por substituição. Inicialmente, isolando $x$

$$2x = 6 - 3y$$
$$x = \frac{6 - 3y}{2}$$

Substituindo $x$ na segunda equação para encontrar $y$

$$2 \left( \frac{6 - 3y}{2} \right) + 2y = 8$$
$$\left( \frac{12 - 6y}{4} \right) + 2y = 8$$
$$\left( \frac{12 - 6y + 8y}{4} \right) = 8$$
$$\left( \frac{12 + 2y}{4} \right) = 8$$
$$12 + 2y = 32$$
$$2y = 20$$
$$y = 10$$

Encontrando $x$

$$x = \frac{6 - 3(10)}{2}$$
$$x = \frac{6 - 30}{2}$$
$$x = \frac{-24}{2}$$
$$x = -12$$

Portanto, a solução é

$$S = (10, -12)$$

```python
# resolvendo com o sympy

equacao1 = 2*_x_ + 3*_y_ - 6
equacao2 = 4*_x_ + 2*_y_ - 8

solve([equacao1, equacao2], [_x_, _y_])
```

```python
def equacao1(x):
    return (6 - 2*x) / 3

def equacao2(x):
    return (8 - 4*x) / 2

x = np.linspace(-5, 5, 100)

y1 = equacao1(x)
y2 = equacao2(x)

plt.plot(x, y1, label='2x + 3y = 6')
plt.plot(x, y2, label='4x + 2y = 8')
plt.scatter(3/2, 1, color='r', zorder=3, label='$(x, y) = (3/2, 1)$')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Sistema linear com uma solução')
plt.legend()

plt.text(-3, 11, '$y = (6 - 2x) / 3$')
plt.text(-4.5, 2.3, '$y = (8 - 4x) / 2$')

plt.show()
```
