```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def equacao_circunferencia(r=1, o=(0, 0)):
    thetas = np.linspace(0, 360, 1000)

    x = [r*math.cos(theta) for theta in thetas]
    y = [r*math.sin(theta) for theta in thetas]

    return x, y
```

```python
x, y = equacao_circunferencia(r=1, o=(0, 0))
o = (0, 0)
```

```python
plt.figure(figsize=(4,4))
plt.plot(x, y, linewidth=.05);
plt.scatter(o[0], o[1]);
```
